**Question 2:** In how many different states do the customers live in?

```python
from os.path import exists
import csv

# open file and init vars
datafile = "./sample_data/customers.csv"
if not exists(datafile):
  raise SystemExit("You should run the first code cell and download the dataset files!")

# customers file is in the sample_data folder and can be opened.
states = set()
with open(datafile, 'r') as fl:
  csvreader = csv.DictReader(fl)

  # evaluate the contents of the file
  for row in csvreader:
    if row["State"]:
      states.add(row["State"].strip())

print(len(states))
```

**Question 3** What is the state with most customers?

```python
import csv
from collections import Counter


# open file and init vars
datafile = "./sample_data/customers.csv"
if not exists(datafile):
  raise SystemExit("You should run the first code cell and download the dataset files!")

# customers file is in the sample_data folder and can be opened.
with open(datafile, 'r', encoding="utf-8") as fl:
  csvreader = csv.DictReader(fl)
  counter = Counter(row["State"] for row in csvreader)
  print(counter.most_common(1))
```

**Question 4** What is the state with the least customers?

```python
import csv
from collections import Counter
from os.path import exists

datafile = "./sample_data/customers.csv"
if not exists(datafile):
    raise SystemExit("Missing file")

with open(datafile, 'r', encoding="utf-8") as fl:
    reader = csv.DictReader(fl)
    counter = Counter(row["State"] for row in reader if row["State"])

# 🔥 encontrar el valor mínimo
min_count = min(counter.values())

# 🔥 obtener todos los estados con ese mínimo
least_common_states = [
    state for state, count in counter.items()
    if count == min_count
]

print("Least common states:", least_common_states)
print("Count:", min_count)
```

**Question #1:** How many unique orders are in the orders.csv file?

**Question #2:** What is the average number of items per order (rounded to two decimal places)?

**Question #3:** What is the highest number of items per order?

**Question #4:** What is the number of orders placed in October 2021?

**Question #5:** Which customer spent the most amount of money in 2021?

**Question #6:** Historically, what is the best month for sales?

Once you get your answers, remember to go back to the course and introduce them in the multiple choice quiz

```python
import csv
from collections import defaultdict, Counter
from datetime import datetime

# open file and init vars
datafile = "./sample_data/orders.csv"
if not exists(datafile):
  raise SystemExit("You should run the first code cell and download the dataset files!")

orders = set()
items_per_order = defaultdict(int)
customer_spending = defaultdict(float)
orders_by_month = Counter()

with open(datafile, 'r', encoding='utf-8') as fl:
    reader = csv.DictReader(fl)

    for row in reader:
        try:
            order_id = row['OrderID']
            customer_id = row['CustomerID']
            date = datetime.strptime(row['Date'], '%Y-%m-%d %H:%M:%S.%f')
            total = float(row['OrderTotal']) if row['OrderTotal'] else 0.0

            # Q1: unique orders
            orders.add(order_id)

            # Q2 & Q3: items per order
            items_per_order[order_id] += 1

            # Q5: spending per customer (solo 2021)
            if date.year == 2021:
                customer_spending[customer_id] += total

            # Q4 & Q6: orders por mes
            month_key = date.strftime('%Y-%m')
            orders_by_month[month_key] += total

        except Exception:
            # Ignorar filas sucias
            continue

# Q1
print("Unique orders:", len(orders))

# Q2
avg_items = sum(items_per_order.values()) / len(items_per_order)
print("Average items per order:", round(avg_items, 2))

# Q3
max_items = max(items_per_order.values())
print("Max items in an order:", max_items)

# Q4 (October 2021)
oct_orders = sum(
    1 for order_id in items_per_order
    if any('2021-10' in k for k in orders_by_month)
)
print("Orders in October 2021:", orders_by_month.get('2021-10', 0))

# Q5
top_customer = max(customer_spending.items(), key=lambda x: x[1])
print("Top customer 2021:", top_customer)

# Q6
best_month = max(orders_by_month.items(), key=lambda x: x[1])
print("Best month:", best_month)
```
